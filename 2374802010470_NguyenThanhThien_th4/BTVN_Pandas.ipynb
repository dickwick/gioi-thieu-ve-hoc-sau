```python
import pandas as pd
import numpy as np
```

```python
# Bài 1
df = pd.read_csv("data/howlongwelive.csv")
df.head(2)
df.tail(2)
df.shape
df.columns
df.describe()
df = df.drop(columns=["Hepatitis B", "Population"])
df["Status"] = df["Status"].map({"Developing": 0, "Developed": 1})
df = df.rename(columns={"thinness 1-19 years": "thinness 10-19 years"})
X = df.drop(columns=["Life expectancy "]).to_numpy()
y = df["Life expectancy "].to_numpy()
X.shape, y.shape
```

```python
# Bài 2
df.isna().sum()
df = df.fillna(df.mean(numeric_only=True))
country_mean = df.groupby("Country")["Life expectancy "].mean()
country_mean.idxmin(), country_mean.min()
country_mean.idxmax(), country_mean.max()
df.groupby("Status")["Life expectancy "].mean()
df_noise = pd.DataFrame({
    "ID": df["Country"],
    "Noise_level": np.random.rand(len(df))
})
df_merged = df.merge(df_noise, left_on="Country", right_on="ID")
df_merged.head()
```
