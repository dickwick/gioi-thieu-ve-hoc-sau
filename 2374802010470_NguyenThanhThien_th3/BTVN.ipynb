pandas and dictionary

```python
# Hệ thống chỉ mục của pandas có cấu trúc tương tự dictionary
#=> Việc tạo một đối tượng pandas từ Dictionary là điều hoàn toàn dễ hiểu
some_population_dict ={'Sai gon' :59,
                       'Đồng Nai':60,
                       'Bình Thuận': 86,
                       'Hà Nội':29}
data_pd = pd.Series(some_population_dict)
data_pd['Sai gon']
```

```python
# Pandas cùng hỗ trợ slicing 
# Lưu ý: Slicing
data_pd['Sai gon':'Hà Nội']
```

Dataframe

DataFrame tương tự mảng 2 chiều, với index hàng và tên cột linh hoạt; về bản chất là tập hợp các Series được căn chỉnh theo index.


```python
some_population_dict ={'Sai gon' :59,
                       'Đồng Nai':60,
                        'Bình Thuận': 86,
                        'Hà Nội':29}
some_area_dict ={'Sai gon' : 51,
                'Đồng Nai': 600,
                'Bình Thuận': 8686,
                'Hà Nội': 33,
                'Vũng Tàu': 72}
state = pd.DataFrame({'population': some_population_dict,
                      'area': some_area_dict})
state
```

indexing

```python
state['Sai gon':'Hà Nội']
```

```python
state['Sai gon':'Hà Nội']['population']
```

```python
state ['population']['Sai gon':'Hà Nội'] #=> Thứ tự không quan trọng
```

```python
state['Sai gon':'Hà Nội']['population']['area']# cột area không tồn tại
```

```python
state['area']
```

```python
state[:]['area']
```
