```python
#BTVN 1:
import torch

x = torch.tensor(2.0, requires_grad=True)
y = 5*x**6 + 6*x**3 - 3*x + 1
y.backward()
print("Giá trị y:", y.item())
print("Độ dốc dy/dx tại x=2:", x.grad.item())
```

```python
#BTVN 2:
import torch
x = torch.tensor(2.0, requires_grad=True)
alpha = 0.1

for i in range(10):
    y = x**3 + 2*x**2 + 5*x + 1
    y.backward()
    with torch.no_grad():
        x -= alpha * x.grad
    x.grad.zero_()
    print(f"Vòng {i+1}: x = {x.item()}")
```

```python
#BTVN 3:
import torch
torch.manual_seed(0)

x = torch.randint(1, 11, (20,)).float()
noise = torch.randn(20)
y = 3*x + 5 + noise

w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
alpha = 0.01

for epoch in range(100):
    y_pred = w*x + b
    loss = ((y_pred - y)**2).mean()
    loss.backward()
    with torch.no_grad():
        w -= alpha * w.grad
        b -= alpha * b.grad
    w.grad.zero_()
    b.grad.zero_()

print("w =", w.item(), "b =", b.item())
```
